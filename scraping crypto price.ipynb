```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
options=webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# fetching bitcoin historical data 
url='https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20200101&end=2020063'
wd=webdriver.Chrome('chromedriver',options=options)
wd.maximize_window()
wd.get(url)

page_num = 0

while wd.find_element_by_xpath("/html/body/div/div/div[1]/div[2]/div/div[3]/div/div/p[1]/button") and page_num<20:
    element=wd.find_element_by_xpath("/html/body/div/div/div[1]/div[2]/div/div[3]/div/div/p[1]/button")
    wd.execute_script("arguments[0].click();", element)
    page_num += 1
    print("getting page number "+str(page_num))
    time.sleep(2)


html=wd.page_source
```

```python
soup=BeautifulSoup(html)
df=pd.read_html(html)
df[0]
```
